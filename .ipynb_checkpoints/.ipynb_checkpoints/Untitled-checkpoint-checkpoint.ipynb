```python
scaler = MinMaxScaler()

x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
smote = SMOTE()

print("Before: ", Counter(y_train))

x_train, y_train = smote.fit_resample(x_train, y_train)

print("After: ", Counter(y_train))
```

```python
def training():
    scores, reports, cms = [], dict(), dict()
    
    for i, j in zip(models, names):
        i.fit(x_train, y_train)
        pred = i.predict(x_test)
        score = accuracy_score(pred, y_test)*100
        report = classification_report(pred, y_test)
        cm = confusion_matrix(pred, y_test)
        scores += [score]
        reports[j] = report
        cms[j] = cm
        
    dd = pd.DataFrame({"score": scores}, index=names)
    dd = dd.sort_values("score", ascending=False)
    dd["score"] = round(dd["score"], 2)
    fig, axes = plt.subplots()
    dd["score"].plot(kind="bar", ax=axes)
    for container in axes.containers:
        axes.bar_label(container)
    axes.set_yticklabels(())
    axes.set_ylabel("")
    axes.set_xlabel("")
    
    plt.show()
    
    index = 0
    
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15, 6))
    
    for i in range(2):
        for j in range(4):
            sns.heatmap(cms[dd.index[index]], annot=True, ax=axes[i][j])
            axes[i][j].set_title("{}: {}%".format(dd.index[index], dd.iloc[index, 0]))
            index += 1
            
    plt.tight_layout()
    plt.show()
    
    for i in dd.index:
        print("*"*30)
        print(i)
        print(reports[i])
        print("\n\n")
```

```python
rfc = RandomForestClassifier()
etc = ExtraTreesClassifier()
gbc = GradientBoostingClassifier()
abc = AdaBoostClassifier()
svc = SVC()
lgr = LogisticRegression()
xgb = XGBClassifier()
lgb = LGBMClassifier()

models = [rfc, etc, gbc, abc,
         svc, lgr, xgb, lgb]

names = ["Random Forest", "Extra Trees", "Gradient Boosting", "Ada Boost",
        "Support Vector Machine", "Logistic Regression", "XGBoost", "LightGBM"]
```

```python
training()
for i, j in zip(models, names):
    i.fit(x_train, y_train)
    with open("file.json", "w") as file:
        file.write(json.dumps(i.get_params()))
```
